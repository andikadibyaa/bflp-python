```python
import pandas as pd

# bikin di dictionary
raw_data = {
    "Nama": ["Andi", "Budi", "Citra", "Deni", "Eka", "Fajar", "Gita", "Hadi"],
    "Usia": [28, 42, 35, 55, 23, 47, 31, 62],
    "Kota": ["Jakarta", "Bandung", "Jakarta", "Surabaya", "Bandung", "Medan", "Jakarta", "Surabaya"],
    "Pekerjaan": ["Karyawan Swasta", "Wiraswasta", "PNS", "Buruh", "Freelancer", "PNS", "Karyawan Swasta", "Pensiunan"],
    "Gaji": [8000000, 15000000, 12000000, 4500000, 6000000, 11000000, 9500000, 5000000],
    "Sisa Kredit": [35000000, 80000000, 20000000, 120000000, 15000000, 45000000, 60000000, 90000000],
    "Tunggakan": [0, 2, 0, 6, 1, 0, 3, 4],
    "PD": [0.03, 0.15, 0.04, 0.35, 0.08, 0.03, 0.12, 0.22],
}

# bikin df
df = pd.DataFrame(raw_data)

# tambah nomor
df.insert(0, "No", range(1, len(df) + 1))

# benerin gaji
df = df.copy()
df["Gaji"] = df["Gaji"].apply(lambda nilai: f"Rp {nilai:,.0f}".replace(",", "."))

print(df)
```

```python
# Ubah PD Budi jadi 0,18
raw_data["PD"][1] = 0.18
print(raw_data)

df = pd.DataFrame(raw_data)
df.insert(0, "No", range(1, len(df) + 1))

# format gaji aman, walau nilainya angka atau string
def format_rupiah(nilai):
    if isinstance(nilai, str):
        bersih = nilai.replace("Rp", "").replace(".", "").replace(",", "").strip()
        nilai = float(bersih) if bersih else 0
    return f"Rp {float(nilai):,.0f}".replace(",", ".")

df["Gaji"] = df["Gaji"].apply(format_rupiah)

print(df)
```

```python
# tambahkan list of dictioanry raw_data 1 kolom yaitu "LGD" semuanya 0,6 
raw_data["LGD"] = []
for _ in range(8):
    raw_data["LGD"].append(0.6)
df = pd.DataFrame(raw_data)
print(df)
```

```python
# tambahkan kolom el=pdxlgdxsisa kredit
df["EL"] = df.apply(lambda row: f"Rp {row['PD'] * row['LGD'] * row['Sisa Kredit']:,.0f}".replace(",", "."), axis=1)
print(f"Expected Loss Andi: Rp {0:,.0f}".format(raw_data['PD'][0] * raw_data['LGD'][0] * raw_data['Sisa Kredit'][0]).replace(",", "."))

```

```python
data = [
    [1, "Andi", 28, "Jakarta", "Karyawan Swasta", 8000000, 350000000, 0, 0.03],
    [2, "Budi", 42, "Bandung", "Wiraswasta", 15000000, 800000000, 2, 0.15],
    [3, "Citra", 35, "Jakarta", "PNS", 12000000, 200000000, 0, 0.04],
    [4, "Deni", 55, "Surabaya", "Buruh", 4500000, 120000000, 6, 0.35],
    [5, "Eka", 23, "Bandung", "Freelancer", 6000000, 150000000, 1, 0.08],
    [6, "Fajar", 47, "Medan", "PNS", 11000000, 450000000, 0, 0.03],
    [7, "Gita", 31, "Jakarta", "Karyawan Swasta", 9500000, 600000000, 3, 0.12],
    [8, "Hadi", 62, "Surabaya", "Pensiunan", 5000000, 900000000, 4, 0.22],
]

kolom = ["No", "Nama", "Usia", "Kota", "Pekerjaan", "Gaji", "Sisa Kredit", "Tunggakan", "PD"]

for k in kolom:
    print(f"{k:<18}", end="")
print()

for baris in data:
    for isi in baris:
        print(f"{str(isi):<18}", end="")
    print()
```

```python
# Hitung EL untuk semua nasabah dan simpan ke dictionary baru dan tampilkan outputnya masing masing nasabah berapa EL nya
el_dict = {}    
print(type(data[0][1]))  # Nama
print(type(data[0][2]))  # Usia 
print(type(data[0][5]))  # Gaji
print(type(data[0][8]))  # PD
print(len(data))  # Jumlah baris  
for baris in data:
    nama = baris[1]
    pd = baris[8]
    lgd = 0.6
    sisa_kredit = baris[6]
    el = pd * lgd * sisa_kredit
    el_dict[nama] = el
    print(f"Expected Loss {nama}: Rp {el:,.0f}".replace(",", "."))
```
