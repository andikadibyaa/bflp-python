```python
import pandas as pd

# bikin di dictionary
raw_data = {
    "Nama": ["Andi", "Budi", "Citra", "Deni", "Eka", "Fajar", "Gita", "Hadi"],
    "Usia": [28, 42, 35, 55, 23, 47, 31, 62],
    "Kota": ["Jakarta", "Bandung", "Jakarta", "Surabaya", "Bandung", "Medan", "Jakarta", "Surabaya"],
    "Pekerjaan": ["Karyawan Swasta", "Wiraswasta", "PNS", "Buruh", "Freelancer", "PNS", "Karyawan Swasta", "Pensiunan"],
    "Gaji": [8000000, 15000000, 12000000, 4500000, 6000000, 11000000, 9500000, 5000000],
    "Sisa Kredit": [35000000, 80000000, 20000000, 120000000, 15000000, 45000000, 60000000, 90000000],
    "Tunggakan": [0, 2, 0, 6, 1, 0, 3, 4],
    "PD": [0.03, 0.15, 0.04, 0.35, 0.08, 0.03, 0.12, 0.22],
}

# bikin df
df = pd.DataFrame(raw_data)

# tambah nomor
df.insert(0, "No", range(1, len(df) + 1))

# benerin gaji
df = df.copy()
df["Gaji"] = df["Gaji"].apply(lambda nilai: f"Rp {nilai:,.0f}".replace(",", "."))

print(df)
```

```python
#Manggil nama-nama di kolom "Nama"
for nama in df["Nama"]:
    print(nama)
```
