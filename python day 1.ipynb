```python
portfolio = [
    {'name': 'Stock A', 'quantity': 100, 'price': 50.0},
    {'name': 'Stock B', 'quantity': 200, 'price': 75.0},
    {'name': 'Stock C', 'quantity': 150, 'price': 60.0},
    {'name': 'Stock D', 'quantity': 250, 'price': 80.0},
    {'name': 'Stock E', 'quantity': 300, 'price': 90.0}
]
```

```python
for stock in portfolio:
    demand = stock['quantity'] * stock['price']
    print(f"{stock['name']:<10} - Demand: {demand:;.1f}")
```
